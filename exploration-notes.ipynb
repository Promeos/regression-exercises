1. Establish Hypotheses
1. Discover patterns
1. Detect anomalies/outliers
1. Calculate summary statistics
1. Graph data to visually identify drivers
> If a graph make you form a question, setup a __hypothesis test__ to test it!
1. Test Hypotheses
- T-tests
- Correlation Tests
- Chi2 Tests


In data exploration, a data scientist looks at the following interactions:
- Continuous v. Continuous Attributes
- Continuous v. Categorical Attributes
- Discrete v. Continuous Attributes
- Discrete v. Categorical Attributes
- Target v. (Continuous, Discrete, and Categorical/Ordinal Attributes)

## Exploration Workflow
---
1. Formulate and write/type your hypotheses
    - Make sure they answer/address your clients questions.
    - Refine your hypothesis
    - Know which statistical test(s) you'll need to perform.
    - What drives the target variable?
    - Which features drive the target variable?
1. Document your findings from each data visualizations
    - Make sure they answer/address your clients questions.
1. Document the results of each hypothesis test
    - Ask a fellow practitioner to review/talk through each test.
1. Document and explicitly state which attributes drive the outcome
    - Explain the relationships between all attributes.
1. Summarize key findings
    - Be brief: Answer the questions.
    - Everything not in the `summary` section should be documented above the summary section.

```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('planets')
```

```python
df.head()
```

```python
df.info()
```

```python
df.method.value_counts()
```

```python
sns.pairplot(df);
```

```python
mask = np.zeros_like(df.corr())
mask[np.triu_indices_from(mask)] = True
mask

with sns.axes_style("white"):

    f, ax = plt.subplots(figsize=(12, 8))
    ax = sns.heatmap(df.corr(),
                     cmap='coolwarm',
                     mask=mask,
                     square=True,
                     annot=True);
```

```python
sns.scatterplot(x=df.distance, y=df.mass, data=df);
```
