```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

url = 'https://gist.githubusercontent.com/zgulde/018f5d601efc79cb741a7aba92f51d29/raw/c08f1c09f605b43e9a354890f9dd68f296e76dc4/students.csv'
df = pd.read_csv(url)
```

```python
df.shape
```

```python
df.info()
```

```python
df['syntax_errors_in_billions'] = df.syntax_errors/1_000_000_000
```

```python
df['cups_of_coffee'] = (df.ml_coffee * 0.033814/8)
```

```python
df['absent_days'] = round(df.p_days_absent * 100, 0)
```

```python
plt.figure(figsize=(8, 4))
plt.scatter(df.syntax_errors_in_billions, df.cups_of_coffee)
plt.title('Number of syntax errors per cup of coffee')
plt.xlabel('Number of syntax errors (in Billions)')
plt.ylabel('Cups of Coffee');
```

```python
plt.figure(figsize=(8, 4))
plt.scatter(df.absent_days, df.cups_of_coffee)
plt.title('Number of days absent per ml of coffee consumed')
plt.xlabel('Days Absent')
plt.ylabel('Cups of Coffee');
```
