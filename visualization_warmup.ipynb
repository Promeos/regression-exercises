```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

url = 'https://gist.githubusercontent.com/zgulde/018f5d601efc79cb741a7aba92f51d29/raw/c08f1c09f605b43e9a354890f9dd68f296e76dc4/students.csv'
df = pd.read_csv(url)
```

```python
df['syntax_errors_in_billions'] = df.syntax_errors/1_000_000_000
```

```python
df['cups_of_coffee'] = (df.ml_coffee * 0.033814/8)
```

```python
# df['absent_days'] = round(df.p_days_absent * 100, 0)
```

```python
plt.figure(figsize=(8, 4))
plt.scatter(df.cups_of_coffee, df.syntax_errors_in_billions)
plt.title('There is a linear correlation between coffee consumption and syntax errors')
plt.xlabel('Coffee Consumption (cups)')
plt.ylabel('# Syntax Errors (in Billions)');
```

```python
def format_y_ticks(tick, position):
    tick = f'{tick:.1%}'
    return tick
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
plt.scatter(df.cups_of_coffee, df.p_days_absent)
plt.title('There is no linear relationship between coffee consumption and absences')
plt.ylabel('Days Absent (%)')
plt.xlabel('Coffee Consumption (cups)')
ax.yaxis.set_major_formatter(mpl.ticker.FuncFormatter(format_y_ticks));
```
